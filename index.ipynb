## Dataset

Run the cell below to import the dataset containing the historical running times for the men's 400m in the Olympic games.

```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
import warnings
from statsmodels.tools.sm_exceptions import ConvergenceWarning

warnings.simplefilter("ignore", ConvergenceWarning)

data = pd.read_csv("winning_400m.csv")
data["year"] = pd.to_datetime(data["year"].astype(str))
data.set_index("year", inplace=True)
data.index = data.index.to_period("Y")
```

```python
# Preview the dataset
data
```

Plot this time series data. 

```python
# Plot the time series
data.plot(figsize=(12,6))
plt.xlabel("Year", fontsize=15)
plt.ylabel("Winning times (in seconds)", fontsize=15)
plt.show()
```

If you plotted the time series correctly, you should notice that it is not stationary. So, difference the data to get a stationary time series. Make sure to remove the missing values.

```python
# Difference the time series
data_diff = data.diff().dropna()
data_diff
```

Use `statsmodels` to plot the ACF and PACF of this differenced time series. 

```python
# Plot the ACF
from statsmodels.graphics.tsaplots import plot_acf

fig, ax = plt.subplots(figsize=(8,3))
plot_acf(data_diff, lags=8, ax=ax);
```

```python
# Plot the PACF
from statsmodels.graphics.tsaplots import plot_pacf

fig, ax = plt.subplots(figsize=(8, 3))
plot_pacf(data_diff, ax=ax, lags=8, method="ywm");
```

Based on the ACF and PACF, fit an ARMA model with the right orders for AR and MA. Feel free to try different models and compare AIC and BIC values, as well as significance values for the parameter estimates. 

```python
"""
AIC takes into account both the likelihood of the model (how well it fits the data) and the number of parameters in the model.
AIC seeks to find the model that best explains the data while penalizing complex models. It balances the trade-off between model fit and model complexity.
Lower AIC values indicate a better trade-off between goodness of fit and model complexity. The model with the lowest AIC is typically preferred.

BIC is similar to AIC in that it measures model goodness of fit and complexity.
BIC also balances model fit and complexity, but it places a higher penalty on models with more parameters than AIC does.
Like AIC, lower BIC values indicate a better trade-off between goodness of fit and model complexity. The model with the lowest BIC is preferred.
"""

"""
differencing (d=1) has already removed any linear trends or seasonality from the data, making it stationary. 
The AR component (p=1) suggests that the current value of the time series is regressed on the immediately preceding value, which can capture some of the autocorrelation in the data. 
The MA component (q=0) suggests that there are no moving average terms, indicating no dependence on past error terms.
"""
# import ARIMA
from statsmodels.tsa.arima.model import ARIMA 

# Fit an ARIMA model (1,0) 
# 1 suggests first order 'p' which is the order of the Autoregressive component. 
# 0 suggests zero order 'd' which is the degree of differencing. 
# q (order of the Moving Average component)

# Instantiate the ARIMA model with the parameters
mod_arma = ARIMA(data_diff, order=(1,0,0))
# Fit the model
res_arma = mod_arma.fit()

# Print summary 
print(res_arma.summary())
```

```python
# Fit an ARMA(2,1) model
mod_arma = ARIMA(data_diff, order=(2, 0, 1))
res_arma = mod_arma.fit()

# Print out summary information on the fit
print(res_arma.summary())
```

```python
# Fit an ARMA(2,2) model
mod_arma = ARIMA(data_diff, order=(2, 0, 2))
res_arma = mod_arma.fit()

# Print out summary information on the fit
print(res_arma.summary())
```
